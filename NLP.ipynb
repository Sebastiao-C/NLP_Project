```python
import numpy as np
import pandas as pd
```

```python
N = 300000
```

```python
lines = list()
with open("500000-lines.txt", "r") as fr:
    a = fr.readline()
    for i in range(N):
        lines.append(a)
        a = fr.readline()
```

```python
with open(f"{N}-lines.txt", "w") as fr:
    for i in range(N):
        fr.write(lines[i])
```

```python
for i,line in enumerate(lines):
    lines[i] = line.removesuffix('\n')
#lines
```

```python
all_words = ' '.join(lines)
```

```python
puncts = ['.', ',', ';', ':', '!', '?', '"', '\'', '\\', '-', '_', '+', '$', '*', '´', '`', '~', '^', 'º', 'ª', '«', '»', '<', '>']
for punct in puncts:
    all_words = all_words.replace(punct, ' ' + punct)#.split(' ')
all_words = all_words.split(' ')
```

```python
all_words[:400]
```

```python
all_words = [x.lower()  for x in all_words if ((x != ' ') and (x != '') and not x.isnumeric() )]
```

```python
all_words
```

```python
min_freq = 10
```

```python
relevant_words = pd.Series(all_words).value_counts()[pd.Series(all_words).value_counts() > min_freq]
```

```python
dict_len = relevant_words.size
```

```python
indices = pd.DataFrame(relevant_words).reset_index()[['index']].reset_index().iloc[:,0]
words = pd.DataFrame(relevant_words).reset_index()[['index']].reset_index().iloc[:,1]
```

```python
ind_to_words_dict = dict(zip(indices, words))
#ind_to_words_dict
```

```python
words_to_ind_dict = dict(zip(words, indices))
#words_to_ind_dict
```

```python
dict_len = len(list(words_to_ind_dict.keys()))
```

```python
dict_len
```

```python
window_size = 8
CO_matrix = np.zeros((dict_len,dict_len))

for i in range(window_size//2, len(all_words)-(window_size//2)):
    if(all_words[i] in words_to_ind_dict.keys()):
        curr_word = all_words[i]
        for k in range(-(window_size//2), (window_size//2)+1):
            if (all_words[i+k] in words_to_ind_dict.keys()):
                ind_curr_word = words_to_ind_dict[curr_word]
                if k < 0:
                    a = (window_size//2)+1 + k
                    CO_matrix[ind_curr_word][words_to_ind_dict[all_words[i+k]]] += a
                elif k > 0:
                    a = (window_size//2)+1 - k
                    CO_matrix[words_to_ind_dict[all_words[i+k]]][ind_curr_word] += a

CO_matrix
```
